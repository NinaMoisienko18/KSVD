**Завдання 1**  

Побудуйте гістограму кількості днів кожного типу погоди в Сіетлі, де:
* верхні краї стовпців — заокруглені;
* порядок по частоті (від найчастішої до найрідшої погоди);
* колір — залежить від типу погоди.

```python
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
```

```python
source = data.seattle_weather()

chart1 = alt.Chart(source).mark_bar(
    cornerRadiusTopLeft=9,
    cornerRadiusTopRight=9
).encode(
    x=alt.X('weather:N', sort='-y'),
    y='count():Q',
    color=alt.Color('weather:N', scale=alt.Scale(range=['navy', 'cornflowerblue', 'blue', 'lavender', 'yellow']))
).properties(
    title="Кількість днів кожного типу погоди"
)

chart1
```

**Завдання 2**  

Побудуйте горизонтальну стовпчикову діаграму врожайності сорту ячменю по місцях у 1931 році:
*   по осі Y — сорти;
*   колір — місце вирощування;
*   дані — лише за 1931 рік;
*   сортування по врожайності.

```python
source = data.barley()

chart2 = alt.Chart(source).transform_filter(
    alt.datum.year == 1931
).mark_bar().encode(
    x='yield:Q',
    y=alt.Y('variety:N', sort='-x'),
    color = alt.Color('site:N', scale = alt.Scale(scheme = 'plasma'))
).properties(
    title='Врожайність ячменю'
)

chart2
```

**Завдання 3**  

Побудуйте гістограму кількості днів певного типу погоди по місяцях:
* по осі X — місяці;
* по осі Y — кількість;
* колір — тип погоди;
* групування: по типу погоди в кожному місяці (груповані стовпчики).

```python
source = data.seattle_weather()

chart3 = alt.Chart(source).mark_bar().encode(
    x=alt.X('month(date):O', title='Місяць'),
    y='count():Q',
    color=alt.Color('weather:N', scale=alt.Scale(range=['navy', 'cornflowerblue', 'blue', 'lavender', 'yellow'])),
    column='weather:N'
).properties(
    title='Кількість днів кожного типу погоди по місяцях'
)

chart3
```

**Завдання 4**  

Побудуйте лінійно-обласну діаграму цін акцій:
* відобразіть лише компанії "GOOG" та "AAPL";
* колір градієнтний залежно від ціни;
* поверх градієнта — лінія.

```python
source = data.stocks()

goog = alt.Chart(source).transform_filter(
    'datum.symbol === "GOOG"'
).mark_area(
    line={'color': 'purple'},
    color=alt.Gradient(
        gradient='linear',
        stops=[
            alt.GradientStop(color='white', offset=0),
            alt.GradientStop(color='purple', offset=1)
        ],
        x1=1, x2=1, y1=1, y2=0
    )
).encode(
    alt.X('date:T'),
    alt.Y('price:Q')
)

aapl = alt.Chart(source).transform_filter(
    'datum.symbol === "AAPL"'
).mark_area(
    line={'color': 'deeppink'},
    color=alt.Gradient(
        gradient='linear',
        stops=[
            alt.GradientStop(color='white', offset=0),
            alt.GradientStop(color='deeppink', offset=1)
        ],
        x1=1, x2=1, y1=1, y2=0
    )
).encode(
    alt.X('date:T'),
    alt.Y('price:Q')
).properties(
    title="Ціни акцій GOOG і AAPL з градієнтом"
)


chart4 = goog + aapl
chart4
```

**Завдання 5**  

Побудуйте потоковий граф (stack='center') рівня безробіття в різних галузях:
* використовуйте data.unemployment_across_industries.url;
* по осі X — дата;
* по осі Y — кількість безробітних;
* колір — назва галузі;
* додайте можливість взаємодії (інтерактивність).

```python
source = data.unemployment_across_industries.url

chart5 = alt.Chart(source).mark_area().encode(
    x=alt.X('yearmonth(date):T', axis=alt.Axis(format='%Y')),
    y=alt.Y('sum(count):Q', stack='center', axis=None),
    color=alt.Color('series:N', scale=alt.Scale(scheme='magma'))
).interactive().properties(
    width=700,
    height=400,
    title='Рівень безробіття в США по галузях'
)

chart5
```

**Завдання 6**  

На основі data.seattle_weather() побудуйте:
* частки кожного типу погоди у вигляді кільцевої діаграми;
* покажіть у центрі кільце середнє значення температури;
* використовуйте mark_arc з innerRadius.



```python
source = data.seattle_weather()

weather_counts = source.groupby('weather').size().reset_index(name='count')

chart6 = alt.Chart(weather_counts).mark_arc(innerRadius=50).encode(
    theta='count:Q',
    color=alt.Color('weather:N', scale=alt.Scale(range=['navy', 'cornflowerblue', 'blue', 'lightgray', 'yellow']))
).properties(
    title="Типи погоди у Сіетлі — кільцева діаграма"
)

chart6
```
