8. Виконайте наступні завдання з пошуку і сортування даних.

Який рейтинг у датасеті найвищий?




```python
max_rating = df_2['user_rating'].max()
max_rating
```

Скільки книг мають такий рейтинг? **(52)**



```python
df_2[df_2['user_rating'] == max_rating]['name'].count()
```

Яка книга має найбільше відгуків?

```python
df_2[df_2['reviews'] == df_2['reviews'].max()]
```

З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ?

```python
max_price = df_2[df_2['year'] == 2010]['price'].max()
df_2[(df_2['year'] == 2010) & (df_2['price'] == max_price)]
```

Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році

```python
df_2[(df_2['year'] == 2012) & (df_2['genre'] == "Fiction")].count()["name"]
```

Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)?

```python
df_2[(df_2['user_rating'] == 4.9) & ((df_2['year'] == 2010) | (df_2['year'] == 2011))]
```
