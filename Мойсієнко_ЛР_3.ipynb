```python
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

uploaded = files.upload()
```

```python
df = pd.read_csv('spotify-2023.csv', encoding='cp1252')
df
```

```python
# df = df[df['streams'] != 'BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3']
# df
df['streams'] = df['streams'].astype(float)
df.info()
```

1. Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.


```python
top_10_songs = df.sort_values('streams', ascending = False).reset_index()
top_10_songs = top_10_songs[['track_name', 'streams']]
top_10_songs
```

```python
plt.figure(figsize=(12,4))
sns.barplot(data = top_10_songs.head(10), y = 'track_name', x = 'streams', palette = 'viridis')

plt.xticks(rotation=45)


plt.title('Топ-10 пісень')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва треку')

plt.show()
```

4. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів.


```python
top_artists = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis')

plt.title('Топ-10 артистів')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Артист')

plt.show()
```
